```python
# Change the excel ordinal time to python datetime
from datetime import datetime, timedelta

def from_excel_ordinal(ordinal, _epoch0=datetime(1899, 12, 31)):
    if ordinal >= 60:
        ordinal -= 1  # Excel leap year bug, 1900 is not a leap year!
    date = (_epoch0 + timedelta(days=ordinal)).replace(microsecond=0)
    return date
```

```python
data['first_review']= data['first_review'].astype(int).apply(from_excel_ordinal)
data['last_review'] = data['last_review'].astype(int).apply(from_excel_ordinal)
data['host_since'] = data['host_since'].astype(int).apply(from_excel_ordinal)
```

```python
def outlier_by(column):
    list_limit = []
    for item in sorted(data[column].unique().tolist()):    
        min_ = data[data[column]==item]['price'].describe()['min']
        max_ = data[data[column]==item]['price'].describe()['max']
        mean_ = data[data[column]==item]['price'].describe()['mean']

        Q1 = data[data[column]==item]['price'].describe()['25%']
        Q3 = data[data[column]==item]['price'].describe()['75%']
        iqr = Q3 - Q1

        lower_limit = Q1 - (1.5 * iqr)
        upper_limit = Q3 + (1.5 * iqr)

        outlier_index = [index for index,value in zip(data[data[column]==item]['index'],data[data[column]==item]['price']) if (value < lower_limit) | (value > upper_limit)]
        not_outlier_index = [index for index,value in zip(data[data[column]==item]['index'],data[data[column]==item]['price']) if (value > lower_limit) & (value < upper_limit)]
        percent_outlier = round(len(outlier_index)/(len(outlier_index)+len(not_outlier_index))*100,2)

        list_limit.append([min_,max_,mean_,lower_limit,upper_limit,outlier_index,not_outlier_index, percent_outlier])
    describe_outlier = pd.DataFrame(data= list_limit,columns=['min','max','mean','lower_limit','upper_limit','outlier_index','not_outlier_index','percent_outlier(%)'],index=sorted(data[column].unique().tolist())).T
    return describe_outlier
```
