#### - change the format of the date, from ordinal time to python datatime

example  : from_excel_ordinal(40000) become datetime.datetime(2009, 7, 6, 0, 0)

```python
# Change the excel ordinal time to python datetime
from datetime import datetime, timedelta

def from_excel_ordinal(ordinal, _epoch0=datetime(1899, 12, 31)):
    if ordinal >= 60:
        ordinal -= 1  # Excel leap year bug, 1900 is not a leap year!
    date = (_epoch0 + timedelta(days=ordinal)).replace(microsecond=0)
    return date
```

```python
calendarUpdated = []
for i in range(len(data)):
    x = data['calendar_updated'].iloc[i].split()
    if 'months' in x:
        calendarUpdated.append(int(x[0])*30)
    if 'weeks' in x:
        calendarUpdated.append(int(x[0])*7)
    elif 'week' in x:
        calendarUpdated.append(int(7))
    elif 'days' in x:
        calendarUpdated.append(int(x[0]))
    elif 'yesterday' in x:
        calendarUpdated.append(int(1))
    elif 'today' in x:
        calendarUpdated.append(int(1))

data['calendar_updated']=calendarUpdated                                        
                                        
```

- Using weighted review to make it fair between review with 1 visit of guest and review with more than 1 visit of guest

weighted review (WR) = (v ÷ (v+m)) × R + (m ÷ (v+m)) × C

    R = average for the listing (mean) = (Rating)
    v = number of rates for the listing
    m = minimum rates required to be listed 
    C = the mean rate across the whole report.

source : http://answers.google.com/answers/threadview/id/507508.html

```python
len(combined_data[combined_data['num_of_visit']<=20])
```

```python
def weighted_rating(df,col):
    m=20
    C = df[col].mean()
    return round((df['num_of_visit']/(df['num_of_visit']+m) * df[col]) + ((m/(df['num_of_visit']+m)) * C),2)
```

```python
def percentile_(dataframe,col,cols):
    for item in [0.9, 0.8, 0.7, 0.6, 0.5, 0.4, 0.3, 0.2, 0.1,0]:
        if item == 0.9 : 
            for i in dataframe[dataframe[col]>=dataframe[col].quantile(item)].index :
                dataframe[cols].iloc[i]='top 10%'
        elif item == 0.8 :
            for i in dataframe[(dataframe[col]<dataframe[col].quantile(item+0.1))&(dataframe[col]>=dataframe[col].quantile(item))].index :
                dataframe[cols].iloc[i]='top 10%-20%'
        elif item == 0.7 : 
            for i in dataframe[(dataframe[col]<dataframe[col].quantile(item+0.1))&(dataframe[col]>=dataframe[col].quantile(item))].index :
                dataframe[cols].iloc[i]='top 20%-30%'
        elif item == 0.6 : 
            for i in dataframe[(dataframe[col]<dataframe[col].quantile(item+0.1))&(dataframe[col]>=dataframe[col].quantile(item))].index :
                dataframe[cols].iloc[i]='top 30%-40%'
        elif item == 0.5 : 
            for i in dataframe[(dataframe[col]<dataframe[col].quantile(item+0.1))&(dataframe[col]>=dataframe[col].quantile(item))].index :
                dataframe[cols].iloc[i]='top 40%-50%'
        elif item == 0.4 : 
            for i in dataframe[(dataframe[col]<dataframe[col].quantile(item+0.1))&(dataframe[col]>=dataframe[col].quantile(item))].index :
                dataframe[cols].iloc[i]='top 50%-60%'
        elif item == 0.3 : 
            for i in dataframe[(dataframe[col]<dataframe[col].quantile(item+0.1))&(dataframe[col]>=dataframe[col].quantile(item))].index :
                dataframe[cols].iloc[i]='top 60%-70%'
        elif item == 0.2 : 
            for i in dataframe[(dataframe[col]<dataframe[col].quantile(item+0.1))&(dataframe[col]>=dataframe[col].quantile(item))].index :
                dataframe[cols].iloc[i]='top 70%-80%'
        elif item == 0.1 : 
            for i in dataframe[(dataframe[col]<dataframe[col].quantile(item+0.1))&(dataframe[col]>=dataframe[col].quantile(item))].index :
                dataframe[cols].iloc[i]='top 80%-90%'
        elif item == 0 :
            for i in dataframe[(dataframe[col]<dataframe[col].quantile(item+0.1))&(dataframe[col]>=dataframe[col].quantile(item))].index :
                dataframe[cols].iloc[i]='top 90%-100%'
    return dataframe[cols]
```

```python
def outlier_by(column):
    list_limit = []
    for item in sorted(combined_data[column].unique().tolist()):    
        min_ = combined_data[combined_data[column]==item]['price'].describe()['min']
        max_ = combined_data[combined_data[column]==item]['price'].describe()['max']
        mean_ = combined_data[combined_data[column]==item]['price'].describe()['mean']

        Q1 = combined_data[combined_data[column]==item]['price'].describe()['25%']
        Q3 = combined_data[combined_data[column]==item]['price'].describe()['75%']
        iqr = Q3 - Q1

        lower_limit = Q1 - (1.5 * iqr)
        upper_limit = Q3 + (1.5 * iqr)

        outlier_index = [index for index,value in zip(combined_data[combined_data[column]==item]['index'],combined_data[combined_data[column]==item]['price']) if (value < lower_limit) | (value > upper_limit)]
        not_outlier_index = [index for index,value in zip(combined_data[combined_data[column]==item]['index'],combined_data[combined_data[column]==item]['price']) if (value > lower_limit) & (value < upper_limit)]
        percent_outlier = round(len(outlier_index)/(len(outlier_index)+len(not_outlier_index))*100,2)

        list_limit.append([min_,max_,mean_,lower_limit,upper_limit,outlier_index,not_outlier_index, percent_outlier])
    describe_outlier = pd.DataFrame(data= list_limit,columns=['min','max','mean','lower_limit','upper_limit','outlier_index','not_outlier_index','percent_outlier(%)'],index=sorted(data[column].unique().tolist())).T
    return describe_outlier
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor,AdaBoostRegressor
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
def regression_model(n_cv):
    RAND_STATE = 101
    score = []
    model_name = ['Lasso','Ridge','DecisionTreeRegressor','RandomForestRegressor','GradientBoostingRegressor','XGBRegressor']
    models = [Lasso,Ridge,DecisionTreeRegressor,RandomForestRegressor,GradientBoostingRegressor,XGBRegressor]
    for item in models:
        rmse = []
        mae = []
        r2_score_test = []
        r2_score_train =[]
        kf = KFold(n_splits=n_cv,shuffle=True)
        for train_index , test_index in kf.split(X,y):
            X_train,X_test,y_train,y_test = X.iloc[train_index],X.iloc[test_index],y.iloc[train_index],y.iloc[test_index]
            model = item(random_state=RAND_STATE)
            model.fit(X_train,y_train)
            prediction = model.predict(X_test)
            prediction_train = model.predict(X_train)
            mae.append(round(mean_absolute_error(y_test,prediction),2))
            rmse.append(round(np.sqrt(mean_squared_error(y_test,prediction)),2))
            r2_score_train.append(round(r2_score(y_train,prediction_train),2))
            r2_score_test.append(round(r2_score(y_test,prediction),2))

        temp = [
            sum(mae)/n_cv,
            sum(rmse)/n_cv,
            sum(r2_score_train)/n_cv,
            sum(r2_score_test)/n_cv
        ]
        score.append(temp)
    
    # for LinearRegressor
    rmse = []
    mae = []
    r2_score_test = []
    r2_score_train =[]
    kf = KFold(n_splits=n_cv,shuffle=True)
    for train_index , test_index in kf.split(X,y):
        X_train,X_test,y_train,y_test = X.iloc[train_index],X.iloc[test_index],y.iloc[train_index],y.iloc[test_index]
        lin = LinearRegression()
        lin.fit(X_train,y_train)
        prediction = lin.predict(X_test)
        prediction_train = lin.predict(X_train)
        mae.append(round(mean_absolute_error(y_test,prediction),2))
        rmse.append(round(np.sqrt(mean_squared_error(y_test,prediction)),2))
        r2_score_train.append(round(r2_score(y_train,prediction_train),2))
        r2_score_test.append(round(r2_score(y_test,prediction),2))

    temp = [
        sum(mae)/n_cv,
        sum(rmse)/n_cv,
        sum(r2_score_train)/n_cv,
        sum(r2_score_test)/n_cv
    ]
    score.append(temp)
    
    model_name.append('LinearRegression')
    
    # for AdaBoostRegressor
    rmse = []
    mae = []
    r2_score_test = []
    r2_score_train =[]
    kf = KFold(n_splits=n_cv,shuffle=True)
    for train_index , test_index in kf.split(X,y):
        X_train,X_test,y_train,y_test = X.iloc[train_index],X.iloc[test_index],y.iloc[train_index],y.iloc[test_index]
        ada = AdaBoostRegressor(base_estimator=DecisionTreeRegressor(random_state=RAND_STATE),random_state=RAND_STATE)
        ada.fit(X_train,y_train)
        prediction = ada.predict(X_test)
        prediction_train = ada.predict(X_train)
        mae.append(round(mean_absolute_error(y_test,prediction),2))
        rmse.append(round(np.sqrt(mean_squared_error(y_test,prediction)),2))
        r2_score_train.append(round(r2_score(y_train,prediction_train),2))
        r2_score_test.append(round(r2_score(y_test,prediction),2))

    temp = [
        sum(mae)/n_cv,
        sum(rmse)/n_cv,
        sum(r2_score_train)/n_cv,
        sum(r2_score_test)/n_cv
    ]
    score.append(temp)
    
    model_name.append('AdaBoostRegressor')
    
    return pd.DataFrame(score,columns=['MAE_test','RMSE_test','R2_Score_Train','R2_Score_Test'],index=model_name)
```

# 4. HyperParameter by using GridSearchCV

<font color='blue'>data_predict_3 is chosen as Features and Target</font>

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
X = data_predict_3.drop(['price','id'],axis=1)
y = data_predict_3['price']
```

```python
RAND_STATE = 101
testSize = 0.15
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=testSize, random_state=RAND_STATE)
```

## i. Lasso

```python
# find optimal alpha with grid search
alpha = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5, 1]
param_grid = dict(alpha=alpha,
                 max_iter = [10,50,100,500,1000])

search_Lasso = GridSearchCV(estimator=Lasso(random_state=RAND_STATE), param_grid=param_grid, scoring='r2', n_jobs=-1)
search_Lasso.fit(X_train, y_train)
print('Best Score: ', search_Lasso.best_score_)
print('Best Params: ', search_Lasso.best_params_)

```

## ii. Ridge

```python
alpha = [0.001, 0.005, 0.01, 0.05, 0.1, 0.5,1,5]
param_grid = dict(alpha=alpha,
                 fit_intercept=[True,False])
search_Ridge = GridSearchCV(estimator=Ridge(random_state=RAND_STATE), param_grid=param_grid, scoring='r2', n_jobs=-1)
search_Ridge.fit(X_train, y_train)
print('Best Score: ', search_Ridge.best_score_)
print('Best Params: ', search_Ridge.best_params_)
```

## iii. DecisionTreeRegressor

```python
parameters = dict(max_depth=[5,10,15,20],
                  min_samples_split=[2,4,8,12],
                  max_leaf_nodes=[8,10,12,16,20],
                  min_samples_leaf=[1,3,5]
                  )
search_Dtree = GridSearchCV(estimator=DecisionTreeRegressor(random_state=RAND_STATE),
                           param_grid=parameters,scoring='r2',n_jobs=-1)
search_Dtree.fit(X_train, y_train)    
print('Best Score: ', search_Dtree.best_score_)
print('Best Params: ', search_Dtree.best_params_)
```

## iv. RandomForestRegressor

```python
parameters = dict(bootstrap=[True,False],
                  n_estimators=[200,300,500,100],
                  max_depth=[5,10,20,30,50],
                  min_samples_split=[2,5],
                  min_samples_leaf=[1,3,5],
                  max_features=["auto","sqrt"]
                  )
search_Rand = GridSearchCV(estimator=RandomForestRegressor(random_state=RAND_STATE),
                           param_grid=parameters,scoring='r2',n_jobs=-1)
search_Rand.fit(X_train, y_train)    
print('Best Score: ', search_Rand.best_score_)
print('Best Params: ', search_Rand.best_params_)
```

## v. GradientBoostingRegressor

```python
parameters = {'learning_rate': [0.01,0.02,0.03],
              'subsample'    : [0.9, 0.5, 0.2],
              'n_estimators' : [100,500,1000],
              'max_depth'    : [4,8,12,16,18,20] 
             }
search_Grad = GridSearchCV(estimator=GradientBoostingRegressor(random_state=RAND_STATE), param_grid = parameters,scoring='r2', n_jobs=-1)
search_Grad.fit(X_train, y_train)    
print('Best Score: ', search_Grad.best_score_)
print('Best Params: ', search_Grad.best_params_)
    
```

## vi. XGBRegressor

```python
parameters = dict(n_estimators=[50,100,200],
              learning_rate=[0.01,0.05,0.08,1],
              gamma=[0,0.3,0.6], 
              subsample=[0.3,0.5,0.7],
              colsample_bytree=[0.05,1], 
              max_depth=[5,8,20] )
search_xgb = GridSearchCV(estimator=XGBRegressor(random_state=RAND_STATE), param_grid = parameters,scoring='r2', n_jobs=-1)
search_xgb.fit(X_train, y_train)    
print('Best Score: ', search_xgb.best_score_)
print('Best Params: ', search_xgb.best_params_)
```

## vii. LinearRegression

```python
parameters = dict(fit_intercept=[True,False],
                  normalize=[True,False],
                  copy_X=[True,False])


search_Lin = GridSearchCV(estimator=LinearRegression(), param_grid = parameters,scoring='r2', n_jobs=-1)
search_Lin.fit(X_train, y_train)    
print('Best Score: ', search_Lin.best_score_)
print('Best Params: ', search_Lin.best_params_)
```

## viii. AdaBoostRegressor

```python
parameters = {
    'base_estimator__min_samples_split' : [i for i in range(2,4)],
    'base_estimator__min_samples_leaf' : [i for i in range(2,4)],
    'n_estimators' : [10,20,50,100,200],
    'learning_rate' :[0.001,0.01,0.05,1, 10]
}
search_ada = GridSearchCV(estimator=AdaBoostRegressor(DecisionTreeRegressor(random_state=RAND_STATE, max_depth=50), random_state=RAND_STATE), param_grid = parameters,scoring='r2', n_jobs=-1)
search_ada.fit(X_train, y_train)    
print('Best Score: ', search_ada.best_score_)
print('Best Params: ', search_ada.best_params_)
```

# 5. Modelling with the best model and the hyperparameter

```python
from sklearn.model_selection import train_test_split
X = data_predict_3.drop(['price','id'],axis=1)
y = data_predict_3['price']
RAND_STATE = 101
testSize = 0.15
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=testSize, random_state=RAND_STATE)
```

### 1. RandomForestRegressor 

```python
Random = RandomForestRegressor(bootstrap= False, max_depth= 50, max_features= 'sqrt', min_samples_leaf= 1, min_samples_split= 2, n_estimators= 300, random_state=RAND_STATE)
Random.fit(X_train,y_train)
```

```python
print('------------------------------------DATA TRAIN---------------------------------')
print(f'MAE : {mean_absolute_error(y_train,Random.predict(X_train))}')
print(f'MSE : {mean_squared_error(y_train,Random.predict(X_train))}')
print(f'RMSE : {np.sqrt(mean_squared_error(y_train,Random.predict(X_train)))}')
print(f'R2 score : {r2_score(y_train,Random.predict(X_train))}')

print('------------------------------------DATA TEST---------------------------------')
print(f'MAE : {mean_absolute_error(y_test,Random.predict(X_test))}')
print(f'MSE : {mean_squared_error(y_test,Random.predict(X_test))}')
print(f'RMSE : {np.sqrt(mean_squared_error(y_test,Random.predict(X_test)))}')
print(f'R2 score : {r2_score(y_test,Random.predict(X_test))}')
```

```python
sns.distplot((y_test-Random.predict(X_test)),kde=False)
plt.title('Residual Plot RandomForestRegression ')
plt.xlabel('Residual')
plt.show()
```
